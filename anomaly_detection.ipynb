```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv("umakant.csv")
df.head(50)
```

```python
df.replace(to_replace ="I/O Timeout",
                 value =0, inplace = True)
df.head(2493)
```

```python
df.replace(to_replace ="Not Connect",
                 value =0, inplace = True)
df.replace(to_replace ="Scan Timeout",
                 value =0, inplace = True)
df.replace(to_replace ="Configure",
                 value =0, inplace = True)
df.replace(to_replace ="Comm Fail",
                 value =0, inplace = True)
df.replace(to_replace ="Unit Down",
                 value =0, inplace = True)
```

```python
df.Cyclone_Inlet_Gas_Temp[292844]
```

```python
df['Cyclone_Inlet_Gas_Temp'] = pd.to_numeric(df['Cyclone_Inlet_Gas_Temp'], downcast = "float")
df['Cyclone_Material_Temp'] = pd.to_numeric(df['Cyclone_Material_Temp'], downcast = "float")
df['Cyclone_Outlet_Gas_draft'] = pd.to_numeric(df['Cyclone_Outlet_Gas_draft'], downcast = "float")
df['Cyclone_cone_draft'] = pd.to_numeric(df['Cyclone_cone_draft'], downcast = "float")
df['Cyclone_Gas_Outlet_Temp'] = pd.to_numeric(df['Cyclone_Gas_Outlet_Temp'], downcast = "float")
df['Cyclone_Inlet_Draft'] = pd.to_numeric(df['Cyclone_Inlet_Draft'], downcast = "float")
```

```python
from sklearn.ensemble import IsolationForest
```

```python
model=IsolationForest(n_estimators=1000,max_samples='auto',contamination=float(0.2),max_features=1.0)
model.fit(df[['Cyclone_Inlet_Gas_Temp','Cyclone_Material_Temp','Cyclone_Outlet_Gas_draft','Cyclone_cone_draft','Cyclone_Gas_Outlet_Temp','Cyclone_Inlet_Draft']])
```

```python
df.Cyclone_Inlet_Gas_Temp.unique()
```

```python
df['anomalies']= model.decision_function(df[['Cyclone_Inlet_Gas_Temp','Cyclone_Material_Temp','Cyclone_Outlet_Gas_draft','Cyclone_cone_draft','Cyclone_Gas_Outlet_Temp','Cyclone_Inlet_Draft']])
df['anomaly'] = model.predict(df[['Cyclone_Inlet_Gas_Temp','Cyclone_Material_Temp','Cyclone_Outlet_Gas_draft','Cyclone_cone_draft','Cyclone_Gas_Outlet_Temp','Cyclone_Inlet_Draft']])
```
